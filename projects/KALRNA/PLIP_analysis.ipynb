```python
import glob
import pandas as pd
from plip.modules.preparation import PDBComplex
```

```python
def get_complex(pdb_file):
    print('Analysing ' + pdb_file)
    comp = PDBComplex()
    comp.load_pdb(pdb_file)
    comp.analyze()
    
    return comp

def get_interaction_set(plip, interaction_search_string):
    keys = []
    for key in plip.interaction_sets.keys():
        if interaction_search_string in key:
            print('found interaction set...')
            print(key)
            keys.append(plip.interaction_sets[key])
    return keys
    
```

```python
pdbs = glob.glob('/Users/res3/michellab/XChem-examples/KALRNA/GDP_frags/GDP_docked/*.pdb')
```

```python
results = {
    'file': [],
    'total_interactions': [],
    'num_hbonds_ldon_strong': [],
    'num_hbonds_pdon_strong': [],
    'num_hbonds_ldon_weak': [],
    'num_hbonds_pdon_weak': [],
    'num_hydrophobic_contacts': [],
    'num_halogen_bonds': [],
    'num_metal_complexes': [],
    'num_saltbridge_lneg': [],
    'num_saltbridge_pneg': [],
    'num_pication_laro': [],
    'num_pication_paro': [],
    'num_pistacking': [],
    'num_water_bridges': [],
    'num_unpaired_hal': [],
    'num_unpaired_hba': [],
    'num_unpaired_hbd': []
}


for pdb in pdbs:
    p = get_complex(pdb)
    interactions = get_interaction_set_keys(p, 'UNL')
    for interaction in interactions:
        if interaction.no_interactions:
            continue
        results['file'].append(pdb.split('/')[-1])
        
        num_hbonds_ldon_strong = len([i.type=='strong' for i in interaction.hbonds_ldon])
        results['num_hbonds_ldon_strong'].append(num_hbonds_ldon)
        
        num_hbonds_ldon_weak = len([i.type=='weak' for i in interaction.hbonds_ldon])
        results['num_hbonds_ldon_weak'].append(num_hbonds_ldon)
        
        num_hbonds_pdon_strong = len([i.type=='strong' for i in interaction.hbonds_pdon])
        results['num_hbonds_pdon_strong'].append(num_hbonds_pdon)
        
        num_hbonds_pdon_weak = len([i.type=='weak' for i in interaction.hbonds_pdon])
        results['num_hbonds_pdon_weak'].append(num_hbonds_pdon)
        
        num_hydrophobic_contacts = len(interaction.hydrophobic_contacts)
        results['num_hydrophobic_contacts'].append(num_hydrophobic_contacts)
        
        num_halogen_bonds = len(interaction.halogen_bonds)
        results['num_halogen_bonds'].append(num_halogen_bonds)
        
        num_metal_complexes = len(interaction.metal_complexes)
        results['num_metal_complexes'].append(num_metal_complexes)
        
        num_saltbridge_lneg = len(interaction.saltbridge_lneg)
        results['num_saltbridge_lneg'].append(num_saltbridge_lneg)
        
        num_saltbridge_pneg = len(interaction.saltbridge_pneg)
        results['num_saltbridge_pneg'].append(num_saltbridge_pneg)
        
        num_pication_laro = len(interaction.pication_laro)
        results['num_pication_laro'].append(num_pication_laro)
        
        num_pication_paro = len(interaction.pication_paro)
        results['num_pication_paro'].append(num_pication_paro)
        
        num_pistacking = len(interaction.pistacking)
        results['num_pistacking'].append(num_pistacking)
        
        num_water_bridges = len(interaction.water_bridges)
        results['num_water_bridges'].append(num_water_bridges)
        
        total = num_halogen_bonds + num_hbonds_ldon_strong + num_hbonds_pdon_strong + num_hydrophobic_contacts 
        + num_metal_complexes + num_pication_laro + num_pication_paro + num_pistacking 
        + num_saltbridge_lneg + num_saltbridge_pneg + num_water_bridges + num_hbonds_ldon_weak + num_hbonds_pdon_weak
        results['total_interactions'].append(total)
        
        results['num_unpaired_hal'].append(interaction.num_unpaired_hal)
        results['num_unpaired_hba'].append(interaction.num_unpaired_hba)
        results['num_unpaired_hbd'].append(interaction.num_unpaired_hbd)
        
```

```python
pd.DataFrame.from_dict(results)
```
